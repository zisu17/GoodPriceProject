```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
import logging
import traceback
```

```python
df = pd.read_csv('./data/서울 음식점 샘플2000(아이디).csv',encoding='utf-8')
```

```python
df
```

```python
url=[]
for i in range(len(df)) :
    base_url = 'https://pcmap.place.naver.com/restaurant/'
    sub_url = df['업소id'][i]+'/home'
    entry_url = base_url + sub_url
    url.append(entry_url)
```

```python
url
```

```python
score=[]
for i in tqdm(range(0,2000)):
    try :
        # 페이지 요청
        html = requests.get(url[i]).content
        # 파싱객체 생성
        soup = BeautifulSoup(html,"html.parser")
        time.sleep(1)
        s=soup.findAll("span",{"class":"_1Y6hi _1A8_M"})[0].em.text
        score.append(s)
    except Exception as e:        
        logging.error(traceback.format_exc())
        score.append('-')
```

```python
df.insert(17,"별점",score)
```

```python
df.to_csv('C:\PythonStudy\project1\data\서울 음식점 샘플2000(아이디)(별점추가).csv',encoding='euc-kr',index=False)
```
